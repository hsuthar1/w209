# 1

Plot a histogram of car types at entrance 1.

````python
def q1():
    # gets count per car type for entrance 1
    temp = df[df["gate-name"]=="entrance1"][["car-type","car-id"]].groupby("car-type").count()
#     print(temp)

    # builds the x domain for our bar chart
    x = range(1,len(temp.index.tolist())+1)
#     print(x)

    # uses the counts dataframe to get labels we can use for the x domain later
    xlabs = temp.index
#     print(xlabs)

    # builds the y range for our bar chart
    y = temp["car-id"]
#     print(y)

    # builds a bar chart
    plt.bar(x,y)
    plt.title("Number of Cars at Entrance 1")
    plt.xticks(x, xlabs)
    plt.xlabel("cart type")
    plt.ylabel("number of cars")
    plt.show()
q1()
````

# 3

Plot the count of daily visits grouped by day for the whole dataset.

a. Hint: The groupby() method of pandas works like this:

```python
g = df.groupby(pd.DatetimeIndex(pd.to_datetime(df["Timestamp"])).date)
```

then you can plot the counts by using g.count()

````python
def q3():
    # gets counts per day
    temp = df.groupby(pd.DatetimeIndex(pd.to_datetime(df["Timestamp"])).date).count()
#     print(temp)
    
    # builds the x domain for our bar chart
    x = range(1,len(temp.index.tolist())+1)
#     print(x)
    
    # because there are so many days, builds a x ticks list that skips over certain days
    xtx = np.arange(min(x), max(x)+1, 30.0)
#     print(xtx)
    
    # and because the x ticks are skipped, have to skip over certain x labels too
    xlabs = temp.index[1::30]
#     print(xlabs)
    
    # builds the y range for our bar chart
    y = temp["car-id"]
#     print(y)
    
    # builds a line chart
    plt.figure(figsize=(24,8))
    plt.plot(x,y)
    plt.title("Count of Daily Visits")
    plt.xticks(xtx, xlabs)
    plt.xlabel("Day")
    plt.ylabel("number of cars")
    plt.show()
q3()
````

# 4

Same as plot (3), but plot a separate line for each car type.

a. Hint 1: It will help to loop over the car-types, and do a group by within each car-type.

```python
for car_type in df["car-type"].unique(): 
  g = df[df["car-type"] == car_type].groupby(pd.DatetimeIndex(df[df["car-type"] == car_type]["Timestamp"]).date)
```

b. Hint 2: Make sure to add a legend!

````python
def q4():
    # gets counts per day
    temp = {}
    for car_type in df["car-type"].unique(): 
        g = df[df["car-type"] == car_type].groupby(pd.DatetimeIndex(df[df["car-type"] == car_type]["Timestamp"]).date).count()["car-type"]
        temp[car_type] = g
#         print(g.head())
#     print(len(temp))
#     print(temp.keys())
#     print(temp.values())
    
    # builds a line chart
    plt.figure(figsize=(24,8))
    for idx, s in temp.items():
        plt.plot(s, label=idx)
    plt.title("Count of Daily Visits per Car Type")
    plt.xlabel("Day")
    plt.ylabel("number of cars")
    plt.legend()
    plt.show()
q4()
````

# 5

Repeat plot 3, but group by the hour.

````python
def q5():
    # gets counts per hour
    temp = {}
    for car_type in df["car-type"].unique(): 
        g = df[df["car-type"] == car_type].groupby(pd.DatetimeIndex(df[df["car-type"] == car_type]["Timestamp"]).hour).count()["car-type"]
        temp[car_type] = g
#         print(g.head())
#     print(len(temp))
#     print(temp.keys())
#     print(temp.values())
    
    # builds a line chart
    plt.figure(figsize=(24,8))
    for idx, s in temp.items():
        plt.plot(s, label=idx)
    plt.title("Count of Hourly Visits per Car Type")
    plt.xlabel("Day")
    plt.ylabel("number of cars")
    plt.legend()
    plt.show()
q5()
````

# 6

Repeat plot 4, grouping by the day of week, and show a stacked bar chart for each car type.

a. Hint 1: pd.DatetimeIndex has the attribute .dayofweek.

b. Hint 2: You can loop over the car types again, and to stack the bars, use plt.bar() with argument bottom=prev (where prev is the height of the bars from before). It may help to start with prev = np.zeros(7) outside of the loop, and add to it at the end of each loop (prev += g.count().Timestamp.values)

````python
def q6():
    # gets counts of daily visits per day of week
    temp = {}
    for car_type in df["car-type"].unique(): 
        g = df[df["car-type"] == car_type].groupby(pd.DatetimeIndex(df[df["car-type"] == car_type]["Timestamp"]).dayofweek).count()["car-type"]
        temp[car_type] = g
#     print(temp)
    
    # for each car type, builds a bar chart
    for idx, s in temp.items():
#         print(s)
        
        # builds the x domain for our bar chart
        x = range(1,len(s.tolist())+2)
#         print(x)
        
        # builds the y range for our bar chart
        y = s.values
        y = np.append(y,[0]) # adds a 0 so that we have room for the legend
#         print(y)
    
        # builds a bar chart
        plt.bar(x, y, label=idx)
        
    plt.title("Number of daily visits per day of week")
    plt.xlabel("cart type")
    plt.ylabel("number of cars")
    plt.legend(loc="right")
    plt.show()
    
q6()
````

````python
N = 5
menMeans = (20, 35, 30, 35, 27)
womenMeans = (25, 32, 34, 20, 25)
menStd = (2, 3, 4, 1, 2)
womenStd = (3, 5, 2, 3, 3)
ind = np.arange(N)    # the x locations for the groups
width = 0.35       # the width of the bars: can also be len(x) sequence

p1 = plt.bar(ind, menMeans, width, yerr=menStd)
p2 = plt.bar(ind, womenMeans, width,
             bottom=menMeans, yerr=womenStd)

plt.ylabel('Scores')
plt.title('Scores by group and gender')
plt.xticks(ind, ('G1', 'G2', 'G3', 'G4', 'G5'))
plt.yticks(np.arange(0, 81, 10))
plt.legend((p1[0], p2[0]), ('Men', 'Women'))

plt.show()
````
