# Altair Tutorial

```python
# https://altair-viz.github.io/getting_started/starting.html
import pandas as pd
data = pd.DataFrame({'a': list('CCCDDDEEE'),
                     'b': [2, 7, 4, 1, 2, 6, 8, 4, 7]})
```

```python
import altair as alt
chart = alt.Chart(data)
alt.Chart(data).mark_point()
```

```python
alt.Chart(data).mark_point().encode(
    x='a',
)
```

```python
alt.Chart(data).mark_point().encode(
    x='a',
    y='b'
)
```

```python
alt.Chart(data).mark_point().encode(
    x='a',
    y='average(b)'
)
```

```python
import altair as alt
```

```python
alt.data_transformers.enable('json')
chart = alt.Chart("https://rawgit.com/john-guerra/d3BrushAndLinkingExample/master/data/Lekagul%20Sensor%20Data.csv").mark_bar().encode(
    y="gate-name:N",
    x="count()"
)

chart
```

```python
alt.data_transformers.enable('json')
chart = alt.Chart("Lekagul Sensor Data.csv").mark_bar().encode(
    y="car-type:N",
    x="count()"
)

chart
```

```python
alt.data_transformers.enable('json')
chart = alt.Chart("Lekagul Sensor Data.csv").mark_bar().encode(
    y="gate-name:N",
    x="count()"
)

chart
```

```python
alt.data_transformers.enable('json')
chart = alt.Chart("Lekagul Sensor Data.csv").mark_line().encode(
    y="count()",
    x="yearmonthdate(Timestamp):T"
).properties(width=900, height=300)

chart
```

```python
alt.data_transformers.enable('json')
chart = alt.Chart("Lekagul Sensor Data.csv").mark_line().encode(
    y="count()"
    ,x="yearmonthdate(Timestamp):T"
    ,color="car-type:N"
).properties(width=900, height=300)

chart
```
